### *Carga y Exploración de los datos*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("PRSA_data_2010.1.1-2014.12.31.csv")
```

```python
df.head()
```

```python
total_datos = len(df)
print("Total de datos antes del proceso: {0}".format(total_datos))
```

```python
for c in df.columns:
    print("{0}\t{1}".format(c, df[c].dtype))
```

#### **Punto 1**  

Se considera que el mejor índice puede ser un timestamp compuesto de los fragmentos de la fecha, pues garantiza que no habrá valores repetidos.

```python
def format_date(year, month, day, hour):
    pattern = "{year}-{month:0>2}-{day:0>2} {hour:0>2}:00"
    return pattern.format(year=year, month=month, day=day, hour=hour)
```

```python
zipped_cols = zip(df.year.values,
                  df.month.values,
                  df.day.values,
                  df.hour.values)
date_format = "%Y-%m-%d %H:%M"
df["timestamp"] = [pd.to_datetime(format_date(y, m, d, h), format=date_format) for y, m, d, h in zipped_cols]
```

```python
df = df.set_index("timestamp")
```

```python
df.head()
```

#### **Punto 2**  

Se eliminan los valores nulos en la columna "pm2.5"

```python
df.dropna(axis=0, subset=['pm2.5'], inplace=True)
```

```python
df.head()
```

```python
datos_nona = len(df)
print('Total de datos después de quitar datos nulos en "pm2.5": {0}'.format(datos_nona))
```

#### **Punto 3**  

Se construye un dataframe de resumen (describe) de las columnas numéricas por año. Luego se hace uno para la única columna categórica.

```python
show_columns = ["pm2.5", "DEWP", "TEMP", "PRES", "Iws", "Is", "Ir"]
describe_by_year = []
for yy in df.year.unique():
    describe_yy = df[df.year==yy][show_columns].describe().transpose()
    describe_yy["year"] = yy
    describe_yy.reset_index(inplace=True)
    describe_yy.set_index(["index", "year"], inplace=True)
    describe_by_year.append(describe_yy)

df_describe = pd.concat(describe_by_year)
```

```python
df_describe.sort_index()
```

```python
cbwd_by_year = []
for yy in df.year.unique():
    cbwd_yy = df[df.year==yy]["cbwd"].describe().to_frame().transpose()
    cbwd_yy["year"] = yy
    cbwd_yy.reset_index(inplace=True)
    cbwd_yy.set_index(["index", "year"], inplace=True)
    cbwd_by_year.append(cbwd_yy)

df_cbwd = pd.concat(cbwd_by_year)
```

```python
df_cbwd.sort_index()
```

Se puede apreciar que el diagrama boxplot para todas las columnas juntas no es muy diciente dadas las diferencias entre sus valores. Es más claro al dividir todas las columnas entre su valor máximo, pero sigue sin ser ideal. Por lo tanto, más adelante se realizan gráficos individuales de las mismas.

```python
cols = ["pm2.5", "DEWP", "TEMP", "PRES", "Iws", "Is", "Ir"]
col_min = {c:min(df[c]) for c in cols}
col_max = {c:max(df[c]) for c in cols}
df_scaled = {c:(df[c].values - col_min[c])/(col_max[c] - col_min[c]) for c in cols}
df_scaled = pd.DataFrame(df_scaled)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
ax[0].set_xlabel("Column")
ax[1].set_xlabel("Column")
ax[0].set_ylabel("Values")
ax[1].set_ylabel("Linear-Scaled Values")
ax[0].boxplot(df[cols].values, labels=cols, showfliers=False)
ax[1].boxplot(df_scaled[cols].values, labels=cols, showfliers=True)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(15, 10))
ax[0,0].boxplot(df["pm2.5"].values, labels=["PM 2.5 - no outliers"], showfliers=False)
ax[0,1].boxplot(df["pm2.5"].values, labels=["PM 2.5 - with outliers"], showfliers=True)
ax[0,2].boxplot(df["DEWP"].values, labels=["Dew Point"], showfliers=False)
ax[0,3].boxplot(df["TEMP"].values, labels=["Temperature C"], showfliers=False)
ax[1,0].boxplot(df["PRES"].values, labels=["Atm. Pressure"], showfliers=False)
ax[1,1].boxplot(df["Iws"].values, labels=["Iws"], showfliers=False)
ax[1,2].boxplot(df["Is"].values, labels=["Is"], showfliers=True)
ax[1,3].boxplot(df["Ir"].values, labels=["Ir"], showfliers=True)
plt.show()
```

#### **Punto 4**  

Se construye un diagrama de dispersión de la concentración de material particulado PM2.5 contra la temperatura. Se aprecia una tendencia que pareciera ser normal (gaussiana), con tendencia a mostrar su pico cerca de la media de temperatura (12 C, aprox.).

```python
print("Temperatura media:", np.mean(df["TEMP"].values))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))
ax.scatter(x=df["TEMP"].values, y=df["pm2.5"].values, color="red")
ax.set_title("PM2.5 concentration vs. Temperature [C]")
ax.set_ylabel("PM 2.5 [um/m^3]")
ax.set_xlabel("Temperature [C]")
plt.show()
```

#### **Punto 5**  

Se calcula la probabilidad de encontrar un valor mayor a 120 en el campo "pm2.5".

```python
prob_total = len(df[df["pm2.5"]>120])/total_datos
prob_nona = len(df[df["pm2.5"]>120])/datos_nona
```

Probabilidad de encontrar un dato mayor a 120 en la columna pm2.5 en el dataset **original**:

```python
print("P(pm2.5 > 120 um/m^3): {0:.2%}".format(prob_total))
```

Probabilidad de encontrar un dato mayor a 120 en la columna pm2.5 en el dataset **tras remover datos nulos**:

```python
print("P(pm2.5 > 120 um/m^3 AND pm2.5 is not None): {0:.2%}".format(prob_nona))
```

#### **Punto 6**  

Se calcula el valor promedio por año para la variable "pm2.5" y se ordena de forma ascendente por valor.

```python
promedio_yy = df.groupby("year")["pm2.5"].mean().sort_values(ascending=True)
```
